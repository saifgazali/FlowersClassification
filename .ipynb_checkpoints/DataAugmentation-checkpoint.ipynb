```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import cv2
import os
import PIL
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential 
```

```python
dataset_url = "https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz"
data_dir = tf.keras.utils.get_file('flower_photos',origin=dataset_url,cache_dir='.',untar=True)
```

```python
data_dir = 'datasets/flower_photos'
```

```python
import pathlib 
data_dir = pathlib.Path(data_dir)
data_dir
```

```python
len(list(data_dir.glob('*/*.jpg')))
```

```python
roses = list(data_dir.glob('roses/*.jpg'))
```

```python
PIL.Image.open(roses[1])
```

```python
flowers_img_dict = {
    'roses': list(data_dir.glob('roses/*.jpg')),
    'daisy': list(data_dir.glob('daisy/*.jpg')),
    'dandelion': list(data_dir.glob('dandelion/*.jpg')),
    'sunflowers': list(data_dir.glob('sunflowers/*.jpg')),
    'tulips': list(data_dir.glob('tulips/*.jpg'))
}
```

```python
flowers_label_dict = {
    'roses': 0,
    'daisy': 1,
    'dandelion': 2,
    'sunflowers': 3,
    'tulips': 4
}
```

```python
flowers_img_dict['roses'][0]
```

```python
img = cv2.imread(str(flowers_img_dict['roses'][0]))
img
```

```python
img.shape
```

```python
# our images are of different dimensions
X,Y = [],[]

for flower_name,image in flowers_img_dict.items():
    print(flower_name)
    print(len(image))
```

```python
#resizing the images
for flower_name,images in flowers_img_dict.items():
    for image in images:
        img = cv2.imread(str(image))
        resized_img = cv2.resize(img,(224,224))
        X.append(resized_img)
        Y.append(flowers_label_dict[flower_name])
```

```python
len(X)
```

```python
len(Y)
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
X_scaled = X/255.0
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X_scaled,Y,test_size=0.2)
```

```python
X_scaled.shape
```

```python
Y.shape
```

```python
Y.dtype
```

```python
X_scaled.dtype
```

```python
Y_train = Y_train.astype(float)
Y_test = Y_test.astype(float)
```

```python
cnn = Sequential()
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=(3,3),strides=2,activation='relu',input_shape=[224,224,3])) # filters - no of filters, kernel_size - size of the feature,strides - after each iteration after how many cell we consider the matrix for multiplication 
cnn.add(tf.keras.layers.MaxPool2D(pool_size=(2,2),strides=2))
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=(3,3),strides=2,activation='relu')) # filters - no of filters, kernel_size - size of the feature,strides - after each iteration after how many cell we consider the matrix for multiplication 
cnn.add(tf.keras.layers.MaxPool2D(pool_size=(2,2),strides=2))
cnn.add(tf.keras.layers.Flatten())
cnn.add(tf.keras.layers.Dense(units=6,activation='relu')) #input layer
cnn.add(tf.keras.layers.Dense(units=6,activation='relu')) # hidden layer
cnn.add((tf.keras.layers.Dense(units=5,activation='softmax'))) 
```

```python
cnn.compile(optimizer='adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(),metrics=['accuracy'])
```

```python
cnn.fit(X_train,Y_train,batch_size=32,epochs=5)
```

```python
cnn.evaluate(X_test,Y_test)
```

### Data Augmentation 

```python
data_augmentation = Sequential([
    layers.experimental.preprocessing.RandomZoom(0.3),
    layers.experimental.preprocessing.RandomHeight(0.3),
    layers.experimental.preprocessing.RandomTranslation(0.5,0.5,fill_mode='reflect'),
    layers.experimental.preprocessing.RandomFlip( mode="horizontal_and_vertical"),
    layers.experimental.preprocessing.RandomCrop(124,124)
])
```

```python
plt.imshow(X_train[0])
```

```python
plt.imshow(data_augmentation(X_train)[0])
```

```python
cnn = Sequential()
cnn.add(data_augmentation)
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=(3,3),strides=2,activation='relu',input_shape=[224,224,3])) # filters - no of filters, kernel_size - size of the feature,strides - after each iteration after how many cell we consider the matrix for multiplication 
cnn.add(tf.keras.layers.MaxPool2D(pool_size=(2,2),strides=2))
cnn.add(tf.keras.layers.Conv2D(filters=32,kernel_size=(3,3),strides=2,activation='relu')) # filters - no of filters, kernel_size - size of the feature,strides - after each iteration after how many cell we consider the matrix for multiplication 
cnn.add(tf.keras.layers.MaxPool2D(pool_size=(2,2),strides=2))
cnn.add(tf.keras.layers.Flatten())
cnn.add(tf.keras.layers.Dense(units=6,activation='relu')) #input layer
cnn.add(tf.keras.layers.Dense(units=6,activation='relu')) # hidden layer
cnn.add((tf.keras.layers.Dense(units=5,activation='softmax'))) 
```

```python
cnn.compile(optimizer='adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(),metrics=['accuracy'])
```

```python
cnn.fit(X_train,Y_train,batch_size=32,epochs=30)
```

```python
cnn.evaluate(X_test,Y_test)
```
